```python
import folium
from folium import Map
from folium import Icon
from folium import Marker
from folium import Circle
from folium.plugins import MiniMap
from branca.element import Figure

stamenterrain = 'stamenterrain'

unaText = '<b>Campus Benjamín Núñez</b>'
sanVicenteText = '<b>San Vicente</b>'

sanVicenteLoc = [9.9632076, -84.0454502]
unaLoc = [9.9697604,-84.1316207]

my_map = Map(location=[9.9783532, -84.0844027], zoom_start=12, tiles=stamenterrain)

Marker(location=sanVicenteLoc, popup=sanVicenteText, icon=Icon(color="orange", icon="ok-sign")).add_to(my_map)
Marker(location=unaLoc, popup=unaText, icon=Icon(color="blue", icon="ok-sign")).add_to(my_map)

Circle(location=unaLoc, color="pink", weight=6, fill_color="red", fill_opacity=0.5, tooltip="Campus Benjamín Núñez").add_to(my_map)
Circle(location=sanVicenteLoc, color="purple", weight=6, fill_color="red", fill_opacity=0.5, tooltip="San Vicente").add_to(my_map)

minimap = MiniMap(title_layer=stamenterrain, position='bottomright')

my_map.save('C:folium.html')

my_map
```

```python
fig = Figure(width=1000, height=500)
fig.add_child(my_map)
fig
```
